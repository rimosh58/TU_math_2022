```python
import numpy as np
import numpy.random as random
import scipy as sp
import pandas as pd
from pandas import Series, DataFrame

import matplotlib.pyplot as plt
import matplotlib as mpl
import seaborn as sns
sns.set()
%matplotlib inline

%precision 3
```

```python
pos_rover = pd.read_csv('01-01_20210910_F9P_Rover.out', delim_whitespace = 1, names = ('TOW', 'fixType', 'gnssFixOK', 'diffSoln', 'carrSoln', 'lon', 'lat', 'height'))
pos_rover[0 : 10]
```

```python
pos_rover_fix3 = pos_rover[pos_rover.fixType == 3]
pos_rover_fix3[0 : 10]
```

```python
pos_rover_coordinates = pos_rover_fix3[['lon', 'lat']]
pos_rover_coordinates[0:10]
```

```python
# MatPlotlib

plt.figure(figsize=(15,15))
plt.scatter(x = pos_rover_coordinates['lon'], y = pos_rover_coordinates['lat'])
#plt.plot(pos_rover_coordinates['lon'], pos_rover_coordinates['lat'], linestyle = 'solid')

plt.title('positioning results')
plt.xlabel('Longitude')
plt.gca().get_xaxis().get_major_formatter().set_useOffset(False)

plt.ylabel('Latitude')
plt.gca().get_yaxis().get_major_formatter().set_useOffset(False)

plt.grid(1)
plt.show()
```

```python
# Plotly

import plotly
import plotly.graph_objs as go

import plotly.io as pio
pio.renderers.default = "colab"

coordinate = go.Scatter(x = pos_rover_coordinates['lon'], y = pos_rover_coordinates['lat'], mode = 'markers')
layout = go.Layout(
    title = 'positioning results',
    xaxis = dict(title="Longtitude"),
    yaxis = dict(title="Latitude")
)

fig = dict(data = [coordinate], layout = layout)
go.Figure(fig).show()
```
